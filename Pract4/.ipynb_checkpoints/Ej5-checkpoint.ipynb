```python
import keras 
import tensorflow as tf
from keras.datasets import cifar10
import numpy as np
from keras.callbacks import History 
import matplotlib.pyplot as plt
```

```python
#mapa logistico del ejercicio
def logistic_map(x):
    return 4*x*(1-x)
```

```python
def randomize(x,y):
    #x matriz de imagenes (imagenes x dimension)
    #y matriz de scores verdaderos (imagenes x categorias)
    indices = np.random.choice(x.shape[0], x.shape[0], replace=False)
    x = x[indices]
    y = y[indices]
    return x,y
```

```python
#funcion que genera los datos
#n_datos es el numero de datos
#xo es la condicion inicial (para t=0)
def generate_data(n_datos,xo):
    x= np.arange(n_datos).astype(float)
    y = np.arange(n_datos).astype(float)
    i = 0
    while i < len(x):
        if i == 0:
            y[i] = xo
        else:
            y[i] = logistic_map(y[i-1])
        i = i+1
    x = x.reshape((x.shape[0],1))
    return x,y
        
```

```python
#genero los datos
n_datos = 500
xo = 0.2
x_total,y_total = generate_data(n_datos,xo)
#hago shuffle de los datos
#x_total, y_total = randomize(x_total,y_total)
#separo en datos de train y test 
x_train = x_total[:375,:]
y_train = y_total[:375]
x_test = x_total[375:,:]
y_test = y_total[375:]
#algunos parametros de la red
n_dim = 1
n_capa1 = 5
n_capa2 = 1
epocas = 200
lr = 1e-6
rg = 1e-5
```

```python
#armo la red neuronal
x = keras.layers.Input(shape=(n_dim,))
l1 = keras.layers.Dense(units=n_capa1,kernel_initializer=keras.initializers.RandomNormal(mean=0.0, stddev=0.01, seed=None),
                        activation='tanh',
                        kernel_regularizer=keras.regularizers.l2(rg))(x)
l2 = keras.layers.Concatenate()([x,l1])
l3 = keras.layers.Dense(units=n_capa2,kernel_initializer=keras.initializers.RandomNormal(mean=0.0, stddev=0.01, seed=None),
                        activation='linear',
                        kernel_regularizer=keras.regularizers.l2(rg))(l2)
model = keras.Model(inputs=x, outputs=l3)
optimizer = keras.optimizers.SGD(learning_rate=lr)
model.compile(optimizer, loss='mse')
history = model.fit(x_train, y_train, epochs=epocas, validation_data=(x_test, y_test),batch_size=100, verbose=2)
```

```python
plt.plot(history.history['val_loss'],label='test data',linewidth=2)
plt.plot(history.history['loss'],label='train data',linewidth=2)
plt.legend()
plt.xlabel('Épocas')
plt.ylabel('Loss')
loss_final_ts = history.history['loss'][-1]
loss_final_tr = history.history['val_loss'][-1]
#plt.text(140,400,'loss train final: '+str(round(loss_final_tr,2))+'\n loss test final: '+str(round(loss_final_ts,2)))
```

```python
y_pr = model.predict(x_test)
plt.scatter(y_pr,y_test)
plt.xlabel('Predicción')
plt.ylabel('Valor real')
x_aux = np.linspace(min(y_test),max(y_test),100)
plt.plot(x_aux,x_aux,c='r',linewidth=2)
```
