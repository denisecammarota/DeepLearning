```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons 
```

```python
#n es la dimension
#p es la cantidad de datasets
#como esta en el enunciado


#inicializo con datos normales con centros random
def inicializar1(n,p,ndatos):
    #genero los datos
    i = 0 #esto es un contador
    muk = np.empty([p,n]) #estos son los mu
    muko = np.empty([p,n]) #estos son los mu originales
    data = np.empty([p,n,ndatos])
    #creo los datos, guardo los muko originales
    #inicializo los muk aleatoriamente
    while i < p:
        j = 0 #esto es otro contador
        while j < n:
            n1 = np.random.randint(100)
            x = np.random.normal(n1,20,ndatos)
            muko[i,j] = n1
            data[i,j] = x
            n1 = np.random.randint(100)
            muk[i,j] = n1
            j = j + 1 
        i = i+1
    data = np.hstack(data)
    data = np.transpose(data)
    return muk,muko,data
```

```python
#n es la dimension
#p es la cantidad de datasets
#como esta en el enunciado


#inicializo con datos normales con centros tal que sean
#clusters disjuntos

def inicializar2(n,p,ndatos):
    #genero los datos
    i = 0 #esto es un contador
    muk = np.empty([p,n]) #estos son los mu
    muko = np.empty([p,n]) #estos son los mu originales
    data = np.empty([p,n,ndatos])
    #creo los datos, guardo los muko originales
    #inicializo los muk aleatoriamente
    while i < p:
        j = 0 #esto es otro contador
        while j < n:
            n1 = np.random.randint(100)
            x = np.random.normal(n1,1,ndatos)
            muko[i,j] = n1
            data[i,j] = x
            n1 = np.random.randint(100)
            muk[i,j] = n1
            j = j + 1 
        i = i+1
    data = np.hstack(data)
    data = np.transpose(data)
    return muk,muko,data
```

```python
#n es la dimension
#p es la cantidad de datasets
#como esta en el enunciado


#inicializo con datos normales con centros tal que sean
#clusters que se solapen completamente 
#condicion inicial particular

def inicializar3(n,p,ndatos):
    #genero los datos
    i = 0 #esto es un contador
    muk = np.empty([p,n]) #estos son los mu
    muko = np.empty([p,n]) #estos son los mu originales
    data = np.empty([p,n,ndatos])
    #creo los datos, guardo los muko originales
    #inicializo los muk aleatoriamente
    while i < p:
        j = 0 #esto es otro contador
        while j < n:
            n1 = np.random.randint(5)
            x = np.random.normal(n1,1,ndatos)
            muko[i,j] = n1
            data[i,j] = x
            n1 = np.random.randint(100)
            muk[i,j] = n1
            j = j + 1 
        i = i+1
    data = np.hstack(data)
    data = np.transpose(data)
    return muk,muko,data
```

```python
#n es la dimension
#p es la cantidad de datasets
#como esta en el enunciado


def inicializar4(n,p,ndatos):
    x,y = make_moons(n_samples=ndatos, noise=0.01)
    muk = np.empty([p,n])
    i = 0
    while i < p:
        muk[i,0] = np.random.rand()
        muk[i,1] = np.random.rand()
        i = i+1
    return muk,muk,x
```

```python
#kmeans version 2, despues de usar las herramientas de numpy
#del problema 3 que vimos en clase

#n es la dimension
#p es la cantidad de datasets
#como esta en el enunciado


def kmeans(n,p,ndatos):
    #primer paso, inicializo
    #"primer paso: partimos de un conjunto de medias y datos"
    muk,muko,data = inicializar4(n,p,ndatos)
    mukv = np.zeros([p,n])
    dist = np.linalg.norm(mukv-muk)
    itmax = 100
    it = 0
    while dist > 0.01 and it != itmax:
        #"segundo paso: asignamos clusters"
        resta = data - muk[:,np.newaxis] #resta datos - medias actuales
        normas = np.linalg.norm(resta,axis=2) #calcula norma de esos vectores
        idmin = np.argmin(normas,axis=0) #saca para cada cual es minimo, codifica clusters
        i = 0 #esto es un contador
        #"tercer paso: se vuelve a calcular la media para cada cluster"
        mukv = np.copy(muk) #aca guardo los means viejos
        plt.figure(it+2)
        while i < p:
            muk[i,:] = np.mean(data[idmin==i],axis=0)
            plt.scatter(data[idmin==i,0],data[idmin==i,1])
            plt.scatter(muk[i,0],muk[i,1],s=200,marker="*")
            i = i+1
        it = it +1 #contamos las iteraciones
        dist = np.linalg.norm(mukv-muk)
    if(it == itmax):
        print("Superadas mas de 100 iteraciones no converge!")
    print(muk)
    
                
kmeans(2,2,100)
```
