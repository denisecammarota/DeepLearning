```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import keras
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler #scaling de los datos entre 0 y 1
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.metrics import make_scorer
from os import listdir
from os.path import isfile, join
plt.style.use('seaborn')
plt.style.use('matplotlibrc.py')
```

```python
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)
```

```python
seed = 7
np.random.seed(seed)
```

```python
mypath = 'Datos'
files = [f for f in listdir(mypath) if isfile(join(mypath, f))] #get all file names of that path
brc = 'Datos/RiesgoBariloche'
df_total = [] #aca guardamos todos los datos
```

# primero los datos de bariloche

```python
#abrimos lo de bariloche
data_brc = pd.read_csv(brc,sep=",",quotechar='"',na_values=[''])
data_brc = data_brc["incidenciaAcum14d"]
df_brc = pd.DataFrame(data_brc)
df_brc = df_brc.to_numpy()
df_brc_1 = df_brc[-20:-10]
df_brc = df_brc[:-20] #todo menos los ultimos datos 
```

```python
total_size = df_brc.shape[0]
train_size = int(0.8*total_size)
test_size = total_size - train_size
train_data = df_brc[:-test_size]
test_data = df_brc[-test_size:]
```

```python
df_total.extend(list(train_data.flatten()))
```

# ahora el resto de los datos para entrenar, formo el scaler

```python
for file in files:
    if file != 'RiesgoBariloche':
        data = pd.read_csv(mypath+str('/')+file,sep=",",quotechar='"',na_values=[''])
        data = data["incidenciaAcum14d"]
        df = pd.DataFrame(data)
        df = df.to_numpy()
        df = df[:-10] 
        df_total.extend(list(df.flatten()))
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
df_total = scaler.fit_transform(np.array(df_total).reshape(-1,1))
tw = 3
```

# ahora armo los datos de train y test

```python
train_data = scaler.transform(train_data.reshape(-1, 1))
x_train_total, y_train_total = create_dataset(train_data,tw)
print(x_train_total.shape,y_train_total.shape)
```

```python
for file in files:
    if file != 'RiesgoBariloche':
        print(file)
        data = pd.read_csv(mypath+str('/')+file,sep=",",quotechar='"',na_values=[''])
        data = data["incidenciaAcum14d"]
        df = pd.DataFrame(data)
        df = df.to_numpy()
        df = df[:-10]
        df = scaler.transform(df.reshape(-1,1))
        x_train, y_train = create_dataset(df,tw)
        x_train_total = np.vstack((x_train_total,x_train))
        y_train_total = np.hstack((y_train_total,y_train))
```

```python
x_train_total = x_train_total.reshape(x_train_total.shape[0],1,x_train_total.shape[1])
```

```python
model = keras.Sequential()
model.add(keras.layers.LSTM(units=32,activation='relu',return_sequences=True,input_shape=(1,tw)))
model.add(keras.layers.LSTM(units=32,activation='relu'))
model.add(keras.layers.Dense(units=1))
model.compile(optimizer='adam',loss=keras.losses.MSE,metrics=['mse']) 
model.summary()
history = model.fit(x_train_total, y_train_total,epochs=500,batch_size=256,verbose=2) 
```

```python
plt.plot(history.history['loss'],label='train')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.legend(fontsize=12)
```

```python
test_data = scaler.transform(test_data.reshape(-1, 1))
x_test_test, y_test_total = create_dataset(test_data,tw)
```

```python
x_test_total = x_test_test
x_test_total = x_test_total.reshape(x_test_total.shape[0],1,x_test_total.shape[1])
y_test_pr = model.predict(x_test_total)
```

```python
y_test_pr = scaler.inverse_transform(y_test_pr.reshape(-1,1))
y_test_total = scaler.inverse_transform(y_test_total.reshape(-1,1))
```

```python
plt.plot(y_test_total)
plt.plot(y_test_pr)
```

```python
print('test mse squared:',mean_squared_error(y_test_total,y_test_pr)) 
```
